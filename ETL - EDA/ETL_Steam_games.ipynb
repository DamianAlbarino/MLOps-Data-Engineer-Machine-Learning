# ETL DE STEAM GAMES

```python
import pandas as pd
```

## Cargamos el Dataset

```python
df = pd.read_json('Datasets OPS\\steam_games.json.gz', compression='gzip', lines=True)
df.head()
```

##### A simple vista podemos ver que tienen bastantes filas completas nulas

```python
#Al ver que el head son filas con todo nulo me deshago de ellas
df.dropna(how='all', inplace=True)
```

```python
#Vemos los nulos por columna
df.isna().sum()
```

## Comienzo de transformacion por columnas

#### Transformacion de columna id

```python
#Vemos los valores unicos que tenemos
df['id'].unique()
```

```python
print(df['id'].value_counts()) #Vemos si hay un valor duplicado
print('-'*50)
print((df['id'].value_counts() > 1).value_counts()) #Cuantos valores duplicados hay
```

```python
#Vemos si hay id nulos
df[df['id'].isna()]

#Vemos que el index 883484 no aporta nada de informacion.
```

```python
#Eliminamos valores duplicados y nulos
df.drop_duplicates(subset='id', inplace=True) 
df.dropna(subset='id', inplace=True)

#Cambiamos el tipo de dato
df['id'] = df['id'].astype(int) 
```

#### Transformacion de columna genres

```python
#Vemos los valores unicos que tenemos
df['genres'].value_counts()
```

```python
# Analisamos los valores nulos
df[df['genres'].isna()]
```

##### Veo que Tags y Genre tienen informacion muy similar que se tienen datos que pueden rellenar los nulos q contiene genre

```python
df['genres'].fillna(df['tags'], inplace=True) #Interpeto que tiene una relacion tags y geners por el contenido que tienen
df.dropna(subset='genres', inplace=True) # borro el resto de los nulos
```

#### Transformacion de columna price

```python
#Vemos los valores unicos que tenemos
df['price'].unique()
```

##### Analisamos cada linea los datos de los que tengan Free en el genero

```python
for i in range(0,len(df['genres'])):
    if df['genres'].iloc[i] != None: 
        for j in df['genres'].iloc[i]:
            if 'Free' in j:
                print(df['genres'].iloc[i])
                print(df['price'].iloc[i])
                print(df['url'].iloc[i])
                print('-'*50)
```

##### Los que tienen valor nulo y en genero o tags Free To Play agregamos valor 0.0

```python
condicion = []
for i in df['genres']:
    if i is None:
        condicion.append(False)
        continue
    elif 'Free to Play' in i:
        condicion.append(True)
    else:
        condicion.append(False)
        
df.loc[(condicion) & (df['price'].isna()),'price'] = 0.0

condicion = []
for i in df['tags']:
    if i is None:
        condicion.append(False)
        continue
    elif 'Free to Play' in i:
        condicion.append(True)
    else:
        condicion.append(False)

df.loc[(condicion) & (df['price'].isna()),'price'] = 0.0
```

##### Los datos nulos decido eliminarlos ya que no los puedo rellenar con cualquier dato, y no suma nada tenerlos vacios.

```python
df.dropna(subset='price', inplace=True)
```

```python
df['price'].unique()
```

##### Redondeamos todos a 2 decimales

```python
datos = []
for i in df['price']:
    if type(i) == float:
        datos.append(round(i,2))
    else:
        datos.append(i)

df['price'] = datos
```

##### Hay casos particulares en el dataframe

```python
#Estos casos partiuclares los ponemos en numero
df.loc[df['price'] == 'Starting at $499.00', 'price'] = 499.00
df.loc[df['price'] == 'Starting at $449.00', 'price'] = 449.00
```

##### Las lineas que tienen en price Free, Demo, Play, Install puedo observar que son gratis, por ende los ponemos en 0.0.

```python
df['price'] = df['price'].astype(str)

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price'].fillna(0.0, inplace=True)

df['price'] = df['price'].astype(float)
```

#### Transformacion de columna relase data

```python
#Vemos los valores unicos que tenemos
for i in df['release_date'].unique():
    print(i)
```

```python
#Vemos los nulos
df[df['release_date'].isna()]
```

##### Nos quedamos con solo el año de la fecha de lanzamiento, este proceso es diferente segun el formato que se encuentran

```python
# Agarramos los años de la fecha de lanzamiento, Formato Año-Mes-Dia
df['release_date'].astype(str)
condicion = []
for i in df['release_date']:
    if i is None:
        condicion.append(False)
        continue
    elif '-' in i:
        condicion.append(True)
    else:
        condicion.append(False)

anio = []
for i in df[condicion]['release_date']:
    anio.append(i.split('-')[0])

df.loc[condicion, 'release_date'] = anio
```

```python
#Pongo en none los datos que no dan informacion
condicion = []
for i in df['release_date']:
    if i is None:
        condicion.append(False)
        continue
    elif 'soon' in i.lower():
        condicion.append(True)
    elif 'TBA' in i:
        condicion.append(True)
    elif 'Beta' in i:
        condicion.append(True)
    else:
        condicion.append(False)

anio = []
for i in df[condicion]['release_date']:
    anio.append(None) 

df.loc[condicion, 'release_date'] = anio

# Formato Dia.Mes.Año
condicion = []
for i in df['release_date']:
    if i is None:
        condicion.append(False)
        continue
    elif '.' in i:
        condicion.append(True)
    else:
        condicion.append(False)

anio = []
for i in df[condicion]['release_date']:
    anio.append(i.split('.')[2])

df.loc[condicion, 'release_date'] = anio
```

##### Correccion de casos particulares

```python
#Arreglamos casos particulares
condicion = []
for i in df['release_date']:
    if i is None:
        condicion.append(False)
        continue
    elif len(i.split()) > 2:
        condicion.append(True)
    else:
        condicion.append(False)

df.loc[condicion, 'release_date'] = '2017'
```

```python
# Agarramos los años de la fecha de lanzamiento
df['release_date'].astype(str)
condicion = []
for i in df['release_date']:
    if i is None:
        condicion.append(False)
        continue
    elif ' ' in i:
        condicion.append(True)
    else:
        condicion.append(False)

anio = []
for i in df[condicion]['release_date'].map(lambda x: str.split(x)):
    if i[1] != 'Q1':                            # Hay un caso particular que tiene Q1 en el inicio.
        anio.append(i[1])
    else:
        anio.append(i[0])


df.loc[condicion, 'release_date'] = anio
```

##### Borramos los nulos y verificamos que haya quedado bien el formato

```python
df.dropna(subset='release_date', inplace=True) #borramos el resto de los nulos
```

```python
df['release_date'].unique()
```

#### Tranformacion de columna developer

```python
# Vemos los valores unicos
for i in df['developer'].unique():
    print(i)
```

```python
# Vemos los nulos
df[df['developer'].isna()]
```

##### Como veo que el publisher tiene algunas similitudes con developer, los nulos los relleno con esta informacion con los que se pueden

```python
# Aplico a los none como los developer aquel que lo publico (publisher)
df.loc[(df['developer'].isna()) & (df['publisher'].notnull()),'developer'] = df[(df['developer'].isna()) & (df['publisher'].notnull())]['publisher']
```

##### Los nulos al no poder rellenarlos con nada los elimino

```python
# Borro el resto de los null
df.dropna(subset='developer', inplace=True)

# pongo la primera letra de todas las palabras en mayuscula
df['developer']= df['developer'].apply(lambda x: x.rstrip().title())
```
